Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10).

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = [10, 10, 10]
b = [0, 0, -10]
```

##### 1. Найдите их сумму. (на листочке)
Сумма a + b = c, где c [10; 10; 0]

##### 2. Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе) 

```python
c = np.add(a, b)
c
```

```python
def vlen(vector: list) -> float:
    """
    vector length calculation
    :param vector: list
    :return: float
    """
    result = 0
    for i in vector:
        result += i**2
    return np.around(np.sqrt(result), 2)
```

```python
vlen(a), vlen(b), vlen(c)
```

### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
# Use parametric equations for plot
fig, ax = plt.subplots(nrows=1, ncols=3)
ax1, ax2, ax3 = ax.flatten()

# circle with R=5
r = 5
t = np.linspace(0, 2*np.pi, 100)
ax1.plot(r*np.cos(t), r*np.sin(t))
ax1.set_xlim([-10, 10])
ax1.set_ylim([-10, 10])

# ellipse with R1=2, R2=3
r1 = 7
r2 = 5
t = np.linspace(0, 2*np.pi, 100)
ax2.plot(r1*np.cos(t), r2*np.sin(t))
ax2.set_xlim([-10, 10])
ax2.set_ylim([-10, 10])

# hyperbola with R1=2, R2=1
r1 = 2
r2 = 1
t = np.linspace(-2, 2, 100)
ax3.plot(r1*np.cosh(t), r2*np.sinh(t), color='steelblue')
ax3.plot(-r1*np.cosh(t), r2*np.sinh(t), color='steelblue')
ax3.set_xlim([-10, 10])
ax3.set_ylim([-5, 5])

fig.set_size_inches(14, 4)
plt.subplots_adjust(wspace=0.25)
```

### 5. Задание (в программе)
##### 1. Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(x, y)
z = 3*x + 5*y - 50
ax.plot_wireframe(x, y, z)
z2 = 3*x + 5*y + 50
ax.plot_wireframe(x, y, z2, color='green')
plt.show()
```

##### 2. Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = figure()
ax = Axes3D(fig)
x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(x, y)
z = -3*x**2 + 5*y**2
ax.plot_surface(x, y, z, color='red')

z2 = -60 * np.sqrt(x**2 + 0.5 * y**2)
ax.plot_surface(x, y, z2)
plt.show()
```
