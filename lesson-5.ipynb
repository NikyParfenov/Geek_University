##### 2.1 Напишите код, проверяющий любую из теорем сложения или умножения вероятности на примере рулетки или подбрасывания монетки.

```python
# Sum(Pi) = 1.0
a = np.array([])
n = 100000
for _ in range(n):
    a = np.append(a, int(np.random.uniform(0, 37)))

unique_counts = np.unique(a, return_counts=True)
# unique_counts[1]/n - массив вероятностей выпадения каждого из значений
print(f'Сумма вероятностей выпадения несовместных независимых чисел рулетки: {np.sum(unique_counts[1]/n)}')
```

##### 3.1 Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей (через биномиальное распределение) и сравните результаты.

```python
k, n = 0, 10000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
x = a + b + c + d
for i in range(0, n):
    if x[i] == 2:
        k = k + 1
#print(a, b, c, d)
#print(x)
print(k, n, k/n)

# Биномиальное распределение
p = len(list(itl.combinations(range(0, n), k))) / 2**n
print(p)
```

##### 5. Дополните код расчетом коэффициента корреляции x и y

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
n = 100
r = 0.7
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/(np.sum(x)*np.sum(x) - n*np.sum(x*x))
b = (np.sum(y) - a*np.sum(x))/n

A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]
cor = ( np.sum((x - np.mean(x)) * (y - np.mean(y))) ) / np.sqrt( np.sum((x - np.mean(x))**2) * np.sum((y - np.mean(y))**2) )

print(a, b)
print(a1, b1)
print(f'Correlation: {cor}')
plt.plot([0, 1], [b, a + b])
plt.show()
```
