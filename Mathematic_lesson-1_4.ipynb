##### 4. Задание (в программе):
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 1001)
k = 1
plt.plot(x, np.cos(k*x))
k = (10 - 1.01) * np.random.random() + 1.01
plt.plot(x, np.cos(k*x))
print(f'Графики y=cos(x) и y=cos({np.around(k, 2)}*x)')

# сделаем красивость - оси по центру, картинку по центру
ax = plt.gca()
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
fig = plt.gcf()
fig.set_size_inches(18, 10)

plt.show()
```
