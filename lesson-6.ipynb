##### 1. Задание (теорема сложения).
Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.

$P(2) + P(5) = 1/6 + 1/6 = 1/3 ~= 0.33(3)$

##### 2. Задание (теорема умножения)
Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.

$P(2)*P(5) = 1/6*1/6 = 1/36 =0.027(7)$

##### 7. Задание** (необязательное) 
Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью все грани этого куба будут белыми?

**Комбинаторикой решить не получилось, но нашел в интернете похожую задачу, которая решалась зависимыми событиями, путем уменьшения исходной выборки. Попробовал адаптировать на это условие.**

Всего 27 одинаковых кусочков, следовательно куб был разрезан по осях на 3х3х3.  
У кусочков может быть 4 варианта раскраски:  
- 8 угловых кусочка с 3-мя белыми сторонами;
- 12 кусочков на гранях с 2-мя белыми сторонами;
- 6 кусочков на сторонах с 1-й белой стороной;
- 1 кусочек внутри куба без белых сторон.

Всего имеется 4 вращения кубика вдоль 3х осей с 2мя направлениями, итого 24 положения кубика:  
- для кубиков с 3-мя белыми сторонами вероятность правильной позиции будет $3/24 = 1/8$;
- для кубиков с 2-мя белыми сторонами вероятность правильной позиции будет $2/24 = 1/12$;
- для кубиков с 1-й белой стороной вероятность правильной позиции будет $4/24 = 1/6$;
- для кубика с черными сторонами вероятность правильной позиции будет $1$.

Начнем вытаскивать кубики, при этом число кубиков в выборке будет уменьшаться:  
1. вероятность вытаскивания черного кубика в правильной позиции: $1/27$;
2. вероятность вытаскивания 6 кубиков с 1-й белой стороной в правильной позиции после п.1:  
$(6/26*1/6)*(5/25*1/6)*...*(1/21*1/6)$;
3. вероятность вытаскивания 12 кубиков с 2-мя белыми сторонаями в правильной позиции после п.1, 2:  
$(12/20*1/12)*(11/19*1/12)*...*(1/9*1/12)$;
4. вероятность вытаскивания 8 кубиков с 3-мя белыми сторонами в правильной позиции после п.1, 2, 3:  
$(8/8*1/8)*(7/7*1/8)*...*(1/1*1/8)$.

Перемножаем все это добро и получаем кошмар математика:  
$1/27 * (6/26*1/6)*(5/25*1/6)*...*(1/21*1/6) * (12/20*1/12)*(11/19*1/12)*...*(1/9*1/12) * (8/8*1/8)*(7/7*1/8)*...*(1/1*1/8) = (1*6!*12!*8!) / (6^6*12^12*8^8*27!)$ - здесь в знаменателе 12 в степени 12, отображение может немного сбоить...
